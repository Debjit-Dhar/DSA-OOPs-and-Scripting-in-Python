```python
import matplotlib.pyplot as pl
x=np.linspace(-5,5,10)
y=np.linspace(-10,10,10)
pl.plot(x,y)
pl.grid(True)
pl.show()
```

```python
def f(x):
    return np.sin(x)
```

```python
x=np.linspace(-10,10,100)
y=f(x)
pl.plot(x,y,marker='o')
pl.grid(True)
pl.xlabel('X')
pl.ylabel('Y')
pl.title('Sine')
pl.show()
```

```python
def f(x,y):
    return 10*(x**0)
```

```python
x=np.linspace(-10,10,20)
y=np.linspace(-10,10,20)
```

```python
x,y=np.meshgrid(x,y)
```

```python
z=f(x,y)
```

```python
ax=pl.axes(projection='3d')
ax.plot_surface(x,y,z)
pl.show()
```
